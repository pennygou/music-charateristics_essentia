#### 1. mp3

```python
import os
import essentia
import essentia.standard
import essentia.streaming
from essentia.standard import *
import essentia.standard as es
import pandas as pd
```

```python
autofile = os.walk('/root/audio test/mp3')
lowlevel_mp3=[]
for path ,dir_list, file_list in autofile:
    for file_name in file_list:
        feas = []
        frames = []
        print('Running autofile:',file_name)
        current_file = os.path.join(path,file_name)
        
        loader = essentia.standard.MonoLoader(filename=current_file)
        audio=loader()
        
        features, features_frames = es.MusicExtractor(lowlevelStats=['mean','var'],
                                              rhythmStats=['mean','var'],
                                              tonalStats=['mean','var'])(current_file)
        
        for i in list(features.descriptorNames()):
            feas.append(features[i])
        lowlevel_mp3.append(feas)
```

```python
df_mp3=pd.DataFrame(lowlevel_mp3,columns=list(features.descriptorNames()))
```

```python
df_mp3
```

```python
#删去多余列
import numpy as np
df_mp3.drop(columns=['metadata.tags.discnumber','metadata.tags.encoding','metadata.tags.tracknumber'], axis=0, inplace=True) 
```
